### Question 1

```python
# Enter your code here.
import matplotlib.pyplot as plt

file = open('./dot-to-dot.txt', 'r')
x = []
y = []

for line in file.readlines():
    line_data = line.split(' ')
    x.append(int(line_data[0]))
    y.append(int(line_data[1]))

file.close()

plt.plot(x, y)
plt.show()
```
