```python
def decrypt_message(ciphertext, key):
    key_length = len(key)
    num_rows = len(ciphertext) // key_length

    # Create a matrix to store the characters
    matrix = [[' ' for _ in range(key_length)] for _ in range(num_rows)]

    # Populate the matrix with the ciphertext
    index = 0
    for col in range(key_length):
        for row in range(num_rows):
            matrix[row][col] = ciphertext[index]
            index += 1

    # Rearrange the columns based on the key
    sorted_indices = sorted(range(key_length), key=lambda k: key[k])

    # Read the columns to obtain the plaintext
    plaintext = ''
    for row in range(num_rows):
        for col in sorted_indices:
            plaintext += matrix[row][col]

    return plaintext

# Encrypted quotes
encrypted_quotes = [
    "npomhhe eeneea rvo'titbeepthr earl i tys ehnlhoot agiennrwv vyd.uheie  so nslat s  ol dottvl",
    'naoensioanosprg tnl til  tieleheedthc yfir ohaiw fn w enzhmogfham eoar oonst  ktir yeosae.s ntl ocs pehvcmt eeviereua',
    'tto at   heubl catedtaimati   dkon mctg es teohop tbh.mioei eepndoso  lg p fbsi alt oacanehutmabd ashe',
    'asnem trhivt  taotosatihftorru .v naeu sera eogtwrm, nly f . no yelea lyeraed a biedena yrheka  driesaheriy sna ba '
]

# Try each key and print the decrypted message
for key in range(1, 27):
    print(f"Decrypted Messages with Key {key}:\n")
    
    for i, quote in enumerate(encrypted_quotes, start=1):
        decrypted_message = decrypt_message(quote, list(range(key)))
        print(f"Quote {i}:\n{decrypted_message}\n")
```
