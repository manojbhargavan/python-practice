```python
def random_walk(steps):
    step_options = [-1, 1]
    step_choice = choice(step_options)
    
    walk = []
    walk.append(step_choice)
    for step in range(1, steps):
        walk.append(walk[step - 1] + choice(step_options))
    
    return walk
```

```python
plt.plot(random_walk(1000))
plt.title('Random Walk')
plt.xlabel('Number of Steps')
plt.ylabel('Distance from Origin')
plt.show()
```

```python
def plot_random_walk(number_of_walks, walk_range):
    
    for p in range(number_of_walks):
        steps = randint(1, walk_range)
        x = list(range(1, steps+1))
        current_walk = random_walk(steps)
        plt.plot(x, current_walk, label= f'Plot Number {p}, Steps: {steps}')
        plt.title('Random Walk')
        plt.xlabel('Number of Steps')
        plt.ylabel('Distance from Origin')
        plt.legend(loc='lower left')

    plt.show()
```

```python
plot_random_walk(5, 100)
```
