```python
import matplotlib.pyplot
import matplotlib.pyplot as plt
from random import choice, randint
%matplotlib inline
plt.rc('figure', figsize=(10,10))
```

```python
def transformation_one(x, y):
    return (0.5 * x, 0.5 * y)

def transformation_two(x, y):
    x, y = transformation_one(x, y)
    return (x + 0.5, y + 0.5)

def transformation_three(x, y):
    x, y = transformation_one(x, y)
    return (x + 1, y)
```

```python
def get_points(number_of_points):
    x = [0]
    y = [0]
    transformation_options = [1, 2, 3]
    
    for i in range(1, number_of_points + 1):
        transformation_func = choice(transformation_options)
        if transformation_func == 1:
            curX, curY = transformation_one(x[i-1], y[i-1])
            x.append(curX)
            y.append(curY)
        elif transformation_func == 2:
            curX, curY = transformation_two(x[i-1], y[i-1])
            x.append(curX)
            y.append(curY)
        else:
            curX, curY = transformation_three(x[i-1], y[i-1])
            x.append(curX)
            y.append(curY)
            
    return (x, y)
```

```python
x, y = get_points(1000000)
```

```python
plt.plot(x, y, 'o')
plt.title('Sierpinski Triangle')
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('Sierpinski Triangle.jpg')
plt.show()
```
