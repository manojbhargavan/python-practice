```python
# Imports
import pandas as pd
import pickle

# Import Data
heart_d = pd.read_csv('../data/heart-disease.csv')

# Features
X = heart_d.drop('target', axis=1)

# Lables
y = heart_d['target']

# Test Train Split (20% test split, 80% train split)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Fit to model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

# Score
print(f'Train Score {clf.score(X_train, y_train)}')
print(f'Test Score {clf.score(X_test, y_test)}')
print()

# Check for a better score
best_fit = -1
best_fit_score = 0
for s in range(10, 101, 10):
    clf = RandomForestClassifier(n_estimators=s)
    clf.fit(X_train, y_train)
    print(f'Estimator: {s}')
    print(f'Train Score {clf.score(X_train, y_train)}')
    print(f'Test Score {clf.score(X_test, y_test)}')
    print('')
    test_score = clf.score(X_test, y_test)
    if best_fit_score < test_score:
        best_fit = s
        best_fit_score = test_score

# 70 seems to the best 
clf = RandomForestClassifier(n_estimators=best_fit)
clf.fit(X_train, y_train)
print('---------Best Fit----------')
print(f'Best Fit Estimator {best_fit}')
print(f'Train Score {clf.score(X_train, y_train)}')
print(f'Test Score {clf.score(X_test, y_test)}')

# Save
import pickle
file_name = 'hdr_modelsv.pkl'
pickle.dump(clf, open(file_name, 'wb'))
print(f'Model Saved to: {file_name}')

# Load
l_model = pickle.load(open(file_name, 'rb'))

# Score Loaded
print(l_model.score(X_train, y_train))
print(l_model.score(X_test, y_test))
```
