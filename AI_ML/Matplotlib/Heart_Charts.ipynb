```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in the Heart disease data
heart_disease_df = pd.read_csv('heart-disease.csv')
heart_disease_df.head()
```

```python
# Set Style for matplotlib
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
# Create the figure
fig, ((ax1), (ax2)) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

# 0. Limits
xlimit_min = heart_disease_df['age'].min()-5
xlimit_max = heart_disease_df['age'].max()+10

# Age vs Heart Rate
# 1. Basic Scatter Plot Compare
heart_rate_scatter_plot = ax1.scatter(x=heart_disease_df["age"],
                                      y=heart_disease_df["thalach"],
                                      c=heart_disease_df["target"],
                                      cmap='coolwarm')
# 2. Add HRs for Normal heart rate (60-100) and mean heart rate
ax1.hlines(y=60,
           xmin=xlimit_min,
           xmax=xlimit_max,
           color='green',
           linestyle='dashdot')
ax1.hlines(y=100,
           xmin=xlimit_min,
           xmax=xlimit_max,
           color='green',
           linestyle='dashdot')
ax1.hlines(y=heart_disease_df['thalach'].mean(),
           xmin=xlimit_min,
           xmax=xlimit_max,
           color='teal',
           linestyle='dashed')
# 3. Add Legends
ax1.legend(*heart_rate_scatter_plot.legend_elements(), title='Has Heart Disease')
# 4. Add Titles and Limits 
ax1.set(title='Age vs Heart Rate - Heart Disease Analysis',
        ylabel='Heart Rate',
        xlim=[xlimit_min, xlimit_max],
        ylim=[50, heart_disease_df['thalach'].max()+15])

# Age vs Cholesterol

# 1. Basic Scatter Plot Compare
cholesterol_scatter_plot = ax2.scatter(x=heart_disease_df['age'],
                                       y=heart_disease_df['chol'],
                                       c=heart_disease_df['target'],
                                       cmap='coolwarm')
# 2. Add HRs for Normal Cholesterol (200-239) and mean Cholesterol
ax2.hlines(y=200,
           xmin=xlimit_min,
           xmax=xlimit_max,
           color='green',
           linestyle='dashdot')
ax2.hlines(y=239,
           xmin=xlimit_min,
           xmax=xlimit_max,
           color='green',
           linestyle='dashdot')
ax2.hlines(y=heart_disease_df['chol'].mean(),
           xmin=xlimit_min,
           xmax=xlimit_max,
           color='teal',
           linestyle='dashed')
# 3. Add Legends
ax2.legend(*cholesterol_scatter_plot.legend_elements(), title='Has Heart Disease')
# 4. Add titles and limits
ax2.set(title='Age vs Cholesterol - Heart Disease Analysis',
        xlabel='Age',
        ylabel='Cholesterol',
        xlim=[xlimit_min,xlimit_max],
        ylim=[heart_disease_df['chol'].min()-20, heart_disease_df['chol'].max()+20])

fig.suptitle('Heart Disease Analysis', fontsize=15, fontweight='bold')

plt.show()
```

```python
# Save Figure
fig.savefig('heart-disease-analysis_v1.png')
```
