```python
# Imports
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)

# Setup Plot
fig, ax = plt.subplots(figsize=(10,5))

ax.plot(x,y)

ax.set(title='Simple Plot', xlabel='X', ylabel='Y')

plt.show()
fig.savefig('images/simple-plot-2.png')
```
