```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
car_sales = pd.read_csv('car-sales.csv')
car_sales.head()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('$','').str.replace(',','').astype('float').astype('int')
car_sales.head()
```

```python
car_sales["Running Total"] = car_sales["Price"].cumsum()
car_sales.head()
```

```python
car_sales["Accounting Date"] = pd.date_range(start="3/1/2024",periods=10)
car_sales.head()
```

```python
# Plot - Pandas wraps the matplotlib API
car_sales.plot(x="Accounting Date", y="Running Total")
plt.show()
```

```python
car_sales.plot(x="Odometer (KM)",y="Running Total",kind="scatter")
plt.show()
```

```python
car_sales_summary = car_sales.groupby('Make').sum(numeric_only=True)
car_sales_summary.reset_index(inplace=True)
car_sales_summary.plot(x="Make",y="Running Total",kind="bar", color='lightgreen')
plt.show()
```

```python
car_sales_summary_avg = car_sales.groupby("Make").mean(numeric_only=True)
car_sales_summary_avg.reset_index(inplace=True)
car_sales_summary_avg.plot(x="Make",y="Odometer (KM)",kind="bar")
plt.show()
```
