```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

#### Line Plot

```python
# Mock Data
# x = np.random.randint(1, 10, 5)
x = np.linspace(1, 100, 100)

# Line Plot
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x**2)
ax.set(title='Line Plot Simple - Power of 2',xlabel='X',ylabel='X^2')
plt.show()
# fig.save('filename')
```

#### Scatter Plot

```python
# Mock Data
x = np.linspace(1, 20, 10)

# Plot
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x, np.log(x))
ax.set(title='Scatter Plot - X vs log(X)', xlabel='X', ylabel='log(X)')
plt.show()
```

```python
# Mock Data
x = np.linspace(1, 20, 100)

# Plot
fig, ax = plt.subplots(figsize=(8,4))
ax.scatter(x, np.sin(x), color='green')
ax.scatter(x, np.cos(x), color='orange')
ax.set(title='Sin/Cos Wave', xlabel='X', ylabel='sin/cos(x)')
plt.show()
```

#### Bar Plot

```python
# Data
sales = {'2019': 180, '2020': 856, '2021': 945, '2022': 560, '2023': 170, '2024': 25 }

# Plot
fig, ax = plt.subplots(figsize=(8,4))
ax.bar(sales.keys(), sales.values())
ax.plot(sales.values(), color='green')
ax.set(title='Medical Kit Sales - Mera Dukan', ylabel='Items Sold')
plt.show()
```

```python
# Plot Horizontal Bar
fix, ax = plt.subplots(figsize=(4,2))
ax.barh(list(sales.keys()), list(sales.values()))
ax.set(title='Medical Kit Sales - Mera Dukan', xlabel='Items Sold')
plt.show()
```

#### Histogram

```python
# Data
x = np.random.randn(1000)

# Plot
fig, ax = plt.subplots(figsize=(8,4))
ax.hist(x)
plt.show()
```

#### Sub Plots Option #1

```python
# Data
x = np.linspace(10, 200, 30)

# Plot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
ax1.plot(x)
ax1.set(title='Linear Function', xlabel='X', ylabel='X')
ax2.scatter(x, np.log(x))
ax2.set(title='Log Function', xlabel='X', ylabel='log(X)')
ax3.scatter(x, np.sin(x))
ax3.set(title='Sine Function', xlabel='X', ylabel='sin(X)')
ax4.scatter(x, np.tan(x))
ax4.set(title='Tan Function', xlabel='X', ylabel='tan(X)')
plt.show()
```

#### Subplots Option #2

```python
# Data
x = np.linspace(10, 200, 30)

# Plot
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# Plot in Row 1, Col 1
ax[0, 0].plot(x, x**2)
ax[0, 0].set(title='Quaratic Function', xlabel='X', ylabel='X^2')

# Plot in Row 1, Col 2
ax[0, 1].scatter(x, np.log(x))
ax[0, 1].set(title='Logarithmic Function', xlabel='X', ylabel='log(X)')

# Plot in Row 2, Col 1
ax[1, 0].scatter(x, np.sin(x))
ax[1, 0].set(title='Sine Wave Function', xlabel='X', ylabel='sin(X)')

# Plot in Row 2, Col 2
ax[1, 1].scatter(x, np.cos(x))
ax[1, 1].set(title='Cosine Wave Function', xlabel='X', ylabel='cos(X)')

plt.show()
```
