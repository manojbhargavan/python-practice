```python
x = np.random.randint(1, 100, 5)
y = np.random.randint(0, 1000, 5)

plt.plot(x, y)
plt.show()
```

```python
# 1st Method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd Method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd Method - use this
x = np.random.randint(1, 100, 5)
y = np.random.randint(0, 100, 5)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

#### Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare Data
np.random.seed(10)
x = np.random.randint(1, 100, 4)
y = np.random.rand(4)

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10, 5))

# 3. Plot Data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel='X', ylabel='Y')

# 5. Show Plot & save
fig.savefig('./images/simple-plot.png')
plt.show()
```

```python
# Data
x = np.linspace(1, 11, 100)
y = np.linspace(1, 11, 100)
```

```python
# Line Plot
fig, ax = plt.subplots(figsize=(4,2))
ax.plot(x, x**2)
ax.set(title='Linear Space Plot', xlabel='X', ylabel='Y')
plt.plot()
plt.show()
```

```python
# Scatter Plot
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x, np.exp(x))
plt.plot()
plt.show()
```

```python
# Another Scatter Plot
x = np.linspace(1, 11, 100)
y = np.linspace(12, 21, 100)

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Bar Plot
car_prices = {'Honda': 10000, 'Tata': 5000, 'Hero': 2500, 'Toyota': 15000}
fig, ax = plt.subplots(figsize=(4,2))
ax.bar(car_prices.keys(), car_prices.values(), color='indigo')
ax.set(title='Sams Cars',xlabel='Car Brands',ylabel='Price($)')
plt.show()
```
